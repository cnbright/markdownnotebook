# 衍射光强公式
$$
I = I_{10}\left(\frac{\sin\alpha}{\alpha}\right)^2\cdot\left(\frac{\sin N\beta}{\sin\beta}\right)^2
$$
其中
$$
\alpha =\frac{\pi a}{\lambda}\sin\theta
\qquad
\beta = \frac{\pi(a+b)}{\lambda}\sin\theta
$$
$N$为狭缝数，$a$为缝宽，$b$为两缝间不透光间隙

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10
a = 4e-5
b = 5e-5
lam = 5e-7
I10 = 1
```

```python
def alpha(theta):
    return np.pi*a*np.sin(theta)/lam

def beta(theta):
    return np.pi*(a+b)*np.sin(theta)/lam

def Itheta(theta, n=10):
    return I10*(np.sin(alpha(theta))/alpha(theta))**2 * (np.sin(n*beta(theta))/np.sin(beta(theta)))**2
```

```python
# theta = np.arange(np.finfo(np.float32).eps,2*np.pi,0.05)
theta = np.arange(-0.01,0.01,0.0001)

result10 = Itheta(theta,2)
result20 = Itheta(theta,4)
result40 = Itheta(theta,5)
result100 = Itheta(theta,6)
```

```python
plt.figure(12)
plt.subplot(221)
plt.plot(theta, result10, 'g', label='N=2')
plt.ylabel(r"$I$")
plt.xlabel(r"$\theta$")
plt.legend()
plt.grid()
plt.subplot(222)
plt.plot(theta, result20, 'g', label='N=4')
plt.ylabel(r"$I$")
plt.xlabel(r"$\theta$")
plt.legend()
plt.grid()
plt.subplot(223)
plt.plot(theta, result40, 'g', label='N=5')
plt.ylabel(r"$I$")
plt.xlabel(r"$\theta$")
plt.legend()
plt.grid()
plt.subplot(224)
plt.plot(theta, result100, 'g', label='N=6')
plt.ylabel(r"$I$")
plt.xlabel(r"$\theta$")
plt.legend()
plt.grid()
plt.ylabel(r"$I$")
plt.xlabel(r"$\theta$")

plt.rcParams['figure.dpi'] = 150
plt.figure(12).tight_layout()#调整整体空白
plt.show()
```
