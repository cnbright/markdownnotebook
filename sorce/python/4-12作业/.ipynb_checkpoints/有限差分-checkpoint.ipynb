```python
import numpy as np
from numba import njit,jit,prange
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

$$
\left\{\begin{array}{l}\frac{\partial u}{\partial t}=\frac{\partial^{2} u}{\partial x^{2}}+\frac{\partial^{2} u}{\partial y^{2}} \quad 0<x<1,0<y<1,0 \leq t \\ u(x, y, 0)=0 \\ u(x, 0, t)=u(x, 1, t)=0 \quad 0<x<1,0 \leq t \\ \frac{\partial u(1, y, t)}{\partial x}=0 \quad 0<y<1,0 \leq t \\ \frac{\partial u(0, y, t)}{\partial x}=0 \quad 0.5<y<1,0 \leq t \\ u(0, y, t)=1 \quad 0<y<=0.5,0 \leq t\end{array}\right.
$$

```python
@njit
def difference_iteration(u,alpha):
    N = u.shape[0]
    M = u.shape[1]
    tn = u.shape[2]


    for k in prange(1,tn-1):
        # 因为x的绝热条件，占了两行，从2开始迭代，有问题不能从2迭代，如果边界特殊，还是得加判断保证边界不变
        for i in prange(1,N-2):
            for j in prange(1,M-1):
                u[i,j,k+1] = (1-4*alpha)*u[i,j,k] + alpha*(u[i+1,j,k] + u[i-1,j,k] + u[i,j+1,k] + u[i,j-1,k])
    
    return u
```

```python
# x网格
N = 20
# y网格
M = 40
# 时间网格
h = 0.05
k=100

alpha = 0.25
tau = alpha*h**2

u = np.zeros([N,M,100])

# 初始条件
u[:,:,0] = 0
# y左右边界条件
u[:,0,:] = 0
u[:,-1,:] = 0

# x右边界条件
u[-1,:,:] = 0
u[-2,:,:] = 0
# x左边界条件
u[0,M//2:,:] = 0
u[1,M//2:,:] = 0

u[0,:M//2,:] = 1

u = difference_iteration(u,alpha)
```

```python
def update(i):
    line = ax.pcolormesh(np.transpose(u[:,:,i]), cmap='Blues')
    text.set_text("t={:.4f}s".format(i*tau))
    return line,

fig,ax = plt.subplots(figsize=(11,8))

line = ax.pcolormesh(np.transpose(u[:,:,0]), cmap='Blues')
text = ax.text(2, 38, "t=0.0000s", fontsize=20)
fig.colorbar(line)
plt.xticks(np.linspace(0,N,5,endpoint=True), np.linspace(0,1,5,endpoint=True), rotation=0, fontsize=15)
plt.yticks(np.linspace(0,M,5,endpoint=True), np.linspace(0,1,5,endpoint=True), rotation=0, fontsize=15)
plt.ylabel(r'$y$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
ani = animation.FuncAnimation(fig, update, interval=10, blit=True, save_count=100)
ani.save("Heat_transfer.gif",writer='pillow')
plt.show()
```
