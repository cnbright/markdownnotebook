```python
import numpy as np
import sympy as sp
from numba import njit,jit,prange
import matplotlib.pyplot as plt
```

```python
NN = 100
t = np.linspace(0,3,NN)
x = np.zeros(NN)
v = np.zeros(NN)
a = np.zeros(NN)

dt = 3./NN

x[0] = 0
v[0] = 0

m = 5e-2
k = 0.05
g = 9.81

'''
一级欧拉近似
'''

for i in range(NN-1):
    x[i+1] = x[i]+v[i]*dt
    v[i+1] = v[i]+(g-k/m*v[i]**2)*dt
```

```python
plt.figure(figsize=(10,5))
plt.plot(t, x)
plt.ylabel(r'$x(m)$',fontsize=40)
plt.xlabel(r'$t(s)$',fontsize=40)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(t, v)
plt.ylabel(r'$v(m/s)$',fontsize=40)
plt.xlabel(r'$t(s)$',fontsize=40)
plt.show()
```

```python
'''
二级欧拉近似
'''

t = np.linspace(0,3,NN)
x = np.zeros(NN)
v = np.zeros(NN)
a = np.zeros(NN)

dt = 3/NN

x[0] = 0
v[0] = 0


for i in range(1,NN):
    x[i] = x[i-1]+v[i-1]*dt
    v[i] = v[i-1]+(g-k/m*v[i-1]**2)*dt
    x[i] = x[i-1]+v[i-1]*dt + 0.5*(v[i]-v[i-1])*dt
    v[i] = v[i-1]+((g-k/m*v[i]**2)+(g-k/m*v[i-1]**2))*dt/2
```

```python
plt.figure(figsize=(10,5))
plt.plot(t, x)
plt.ylabel(r'$x(m)$',fontsize=40)
plt.xlabel(r'$t(s)$',fontsize=40)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(t, v)
plt.ylabel(r'$v(m/s)$',fontsize=40)
plt.xlabel(r'$t(s)$',fontsize=40)
plt.show()
```

```python
'''
给定一组一阶微分方程
一阶微分 = xxx
二阶微分 = xxx

输入，
一阶微分的系数
二阶微分的系数
'''
    
```

```python
class euler_approximation:
    def __init__(self, NN, ts, te, x0, v0):
        self.NN = NN
        self.t = np.linspace(ts,te,NN)
        self.dt = (self.t[-1]-self.t[0])/self.NN
        self.x = np.zeros(NN)
        self.v = np.zeros(NN)
        self.x[0] = x0
        self.v[0] = v0
        
    
    def first_order_differential(self,i):
        # 一阶微分占位
        return self.v[i]
    
    def second_order_differential(self,i):
        # 二阶微分占位
        return 0
    
    def first_approximation(self):
        for i in range(self.NN-1):
            self.x[i+1] = self.x[i]+self.first_order_differential(i)*self.dt
            self.v[i+1] = self.v[i]+self.second_order_differential(i)*self.dt
    
    def second_approximation(self):
        for i in range(self.NN-1):
            self.x[i+1] = self.x[i]+self.first_order_differential(i)*self.dt
            self.v[i+1] = self.v[i]+self.second_order_differential(i)*self.dt
            self.x[i+1] = self.x[i]+self.v[i]*self.dt + 0.5*(self.v[i+1]-self.v[i])*self.dt
            self.v[i+1] = self.v[i]+(self.second_order_differential(i+1)+self.second_order_differential(i))*self.dt/2
```

```python
m = 5e-2
k = 0.05
g = 9.81

class exp1(euler_approximation):
    # 一阶微分
    def first_order_differential(self,i):
        return self.v[i]
    
    # 二阶微分
    def second_order_differential(self,i):
        return g-k/m*self.v[i]**2

exp11 = exp1(100, 0., 3., 0, 0)
exp11.first_approximation()

exp12 = exp1(100, 0., 3., 0, 0)
exp12.second_approximation()
```

```python
plt.figure(figsize=(10,5))
plt.plot(exp11.t, exp11.x)
plt.plot(exp12.t, exp12.x)
plt.legend(["first approximation","second approximation"])
plt.ylabel(r'$x(m)$',fontsize=40)
plt.xlabel(r'$t(s)$',fontsize=40)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(exp11.t, exp11.v)
plt.plot(exp12.t, exp12.v)
plt.legend(["first approximation","second approximation"])
plt.ylabel(r'$v(m/s)$',fontsize=40)
plt.xlabel(r'$t(s)$',fontsize=40)
plt.show()
```
